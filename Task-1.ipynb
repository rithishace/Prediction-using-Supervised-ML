### Importing Libraries

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
%matplotlib inline
from sklearn.metrics import accuracy_score
import sklearn.metrics as sm
```

### Reading Dataset from the given url

```python

url = "https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv"
dataset = pd.read_csv(url)
print("Data imported successfully")
```

```python
dataset.head()
```

#### Here, x will be our independent data i.e., Hours
#### And, y will be dependent data ie., Scores
#### Simply put, y is the outcome and x is the cause


```python
x=dataset.iloc[:,0].values.reshape(-1,1)
y=dataset.iloc[:,1].values.reshape(-1,1)
```

### Visualizing the dataset

```python
plt.scatter(x,y)
plt.title('Hours vs Percentage')  
plt.xlabel('Hours Studied')  
plt.ylabel('Percentage Score')  
plt.show
```

Spliting dataset into test and training set, usually 20% of data is taken for testing.

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state=0)
```

# Training Linear Regression Model

```python
lm=LinearRegression()
lm.fit(x_train,y_train)
```

Predicting the outcome.

```python
y_pred=lm.predict(x_test)
```

Plotting the regression line

```python
plt.scatter(x_test,y_test,color="green") # Plot a graph with x_test vs y_test
plt.plot(x_train,lm.predict(x_train),color="red",linewidth=3) # Regressior line showing
plt.title('Regression(Test Set)')
plt.xlabel('Hours')
plt.ylabel('Percentage')
plt.show()

plt.scatter(x_train,y_train,color="blue")  # Plot a graph with x_train vs y_train
plt.plot(x_train,lm.predict(x_train),color="red",linewidth=3) # Regressior line showing
plt.title('Regression(training Set)')
plt.xlabel('Hours')
plt.ylabel('Percentage')
plt.show()
```

## Displaying the Predicted Output

```python
check= pd.DataFrame(x_test,columns=['Hours'])
```

```python
check['actual percentage']=y_test
check['predicted percentage']=y_pred
```

```python
check
```

### Performance metrics


```python
print("Mean absolute error =", round(sm.mean_absolute_error(y_test, y_pred), 2)) 
print("Mean squared error =", round(sm.mean_squared_error(y_test, y_pred), 2)) 
print("Median absolute error =", round(sm.median_absolute_error(y_test, y_pred), 2)) 
print("Explain variance score =", round(sm.explained_variance_score(y_test, y_pred), 2)) 
print("R2 score =", round(sm.r2_score(y_test, y_pred), 2))
print("the score for the regression model is:",lm.score(x_train , y_train))
```
